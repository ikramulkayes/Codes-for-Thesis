```python
import pandas as pd

# Load the CSV file
file_path = 'processed_0to1100.csv'  # Replace with the correct path if necessary
data = pd.read_csv(file_path)

# Function to classify rows based on matching parametric and contextual answers
def classify_row(row):
    # Extract the parametric and contextual answers from the 'Answer' column
    parametric_answer = row['Answer'].split("parametric answer:")[1].split("\ncontextual answer:")[0].strip()
    contextual_answer = row['Answer'].split("contextual answer:")[1].strip()
    
    # Check if they match and classify the row
    return 'Parametric_data' if parametric_answer == contextual_answer else 'Counter_factual_data'

# Apply classification to each row and add a new column 'Category'
data['Category'] = data.apply(classify_row, axis=1)

# Split the data based on the 'Category' column
parametric_data = data[data['Category'] == 'Parametric_data'].drop(columns=['Category'])
counter_factual_data = data[data['Category'] == 'Counter_factual_data'].drop(columns=['Category'])

# Save the split datasets into separate CSV files
parametric_data.to_csv('Parametric_data_deepseek.csv', index=False)
counter_factual_data.to_csv('Counter_factual_data_deepseek.csv', index=False)

print("CSV files saved successfully: 'Parametric_data.csv' and 'Counter_factual_data.csv'")
```
